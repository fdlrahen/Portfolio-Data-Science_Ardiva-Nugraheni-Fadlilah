```python
#collapse-hide
# Reference: https://gist.github.com/slowkow/7a7f61f495e3dbb7e3d767f97bd7304b

def remove_emoji(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

Hashtags/Mentions

```python
def remove_tags_mentions(text):
    pattern = re.compile(r'(@\S+|#\S+)')
    return pattern.sub('', text)
```

```python
df.cleaning = df.cleaning.apply(lambda x: remove_tags_mentions(x))
df.head()
```

Punctuations

```python
PUNCTUATIONS = string.punctuation

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', PUNCTUATIONS))
```

```python
df.cleaning = df["cleaning"].apply(lambda text: remove_punctuation(text))
df.head()
```

Extra Whitespaces

```python
def remove_whitespaces(text):
    return " ".join(text.split())
```

```python
df.cleaning = df.cleaning.apply(lambda x: remove_whitespaces(x))
```

Delete All Symbols

```python
df['cleaning'] = df['cleaning'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
df.head()
```
